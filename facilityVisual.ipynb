```python
import folium
import pandas as pd
from folium import Marker
```

```python
data = pd.read_csv("/content/drive/MyDrive/3학년 2학기/집중교육1/프젝/facilityClustered.csv")
print(data.shape)
data.head()
```

```python
# 교통약자 이용시설 지도 위에 매핑
m = folium.Map(location=[37.511877, 127.059505])
for _, row in data.iterrows():
    Marker(location = [row['WGS84위도'], row['WGS84경도']],
           popup=row['기관명'],
           icon=folium.Icon(color='blue')
          ).add_to(m)
m
```

```python
# Anomaly Detection
anomaly = pd.read_csv("/content/drive/MyDrive/3학년 2학기/집중교육1/프젝/anomaly.csv")

for _, row in anomaly.iterrows():
    Marker(location = [row['WGS84위도'], row['WGS84경도']],
           popup=row['기관명'],
           icon=folium.Icon(color='red')
          ).add_to(m)
m
```

```python
# KMeans++ clustering
m2 = folium.Map(location=[37.511877, 127.059505])
for i, row in data.iterrows():
    if(data.loc[i]['label']) == 0:
        c='red'
    elif(data.loc[i]['label']) == 1:
        c='orange'
    elif(data.loc[i]['label']) == 2:
        c='green'
    elif(data.loc[i]['label']) == 3:
        c='blue'
    Marker(location = [row['WGS84위도'], row['WGS84경도']],
           popup=row['기관명'],
           icon=folium.Icon(color=c,icon='star')
          ).add_to(m2)
m2
```

```python
from folium.plugins import MarkerCluster
```

```python
# visualization clustering
m3 = folium.Map(location=[37.511877, 127.059505])

mc = MarkerCluster()

for i, row in data.iterrows():
    if(data.loc[i]['label']) == 0:
        c='red'
    elif(data.loc[i]['label']) == 1:
        c='orange'
    elif(data.loc[i]['label']) == 2:
        c='green'
    elif(data.loc[i]['label']) == 3:
        c='blue'
    
    mc.add_child(    
        Marker(location = [row['WGS84위도'], row['WGS84경도']],
               popup=row['기관명'],
               icon=folium.Icon(color=c,icon='star')
              )
    )

m3.add_child(mc)

# display the map
m3
```

```python
# 선정한 버스 노선 지도 위에 매핑
route = pd.read_csv("/content/drive/MyDrive/3학년 2학기/집중교육1/프젝/229000042.csv")
m4=folium.Map(location=[37.511877, 127.059505])
route = route[['y', 'x']]
route_list = route.values.tolist()

folium.PolyLine(locations=route_list, tooltip='PolyLine', color='red').add_to(m2)
m2
```
